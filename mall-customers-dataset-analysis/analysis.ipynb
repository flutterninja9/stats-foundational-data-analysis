```python
import kagglehub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Download latest version
path = kagglehub.dataset_download("kandij/mall-customers", path = 'Mall_Customers.csv')
df = pd.read_csv(path)
df.head()
```

```python
# Check for missing values

df.columns[df.isna().any()].tolist()
```

```python
# 1. Identify outliers in the annual income feature (Mall Customers dataset) using IQR.
# 2. Find and remove outliers in spending score based on age & income (Mall Customers dataset).
```

```python
sns.histplot(df['Annual Income (k$)'], kde = True, bins = 30)
plt.show()
```

```python
plt.boxplot(df['Annual Income (k$)'], orientation='horizontal')
plt.show()
```

```python
# Calculate outliers using IQR
Q1 = df['Annual Income (k$)'].quantile(0.25)
Q3 = df['Annual Income (k$)'].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_income = df[(df['Annual Income (k$)'] > lower_bound) & (df['Annual Income (k$)'] < upper_bound)]
filtered_income
```

```python
sns.histplot(filtered_income['Annual Income (k$)'], kde = True, bins = 30)
plt.show()
```

```python
plt.boxplot(filtered_income['Annual Income (k$)'], orientation='horizontal')
plt.show()
```
