```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
## Scatterplot

df = pd.read_csv('data/dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(12,8), dpi=500)
sns.scatterplot(x=df['salary'], y=df['sales'], hue = df['level of education'], alpha = 0.8, style=df['level of education'], s=100)
plt.show()
```

```python
df.head()
```

```python
sns.boxplot(data=df, x='sales', hue='level of education')
plt.show()
```

```python
sns.histplot(data=df, x='salary', kde=True)
```

```python
sns.rugplot(data = df, x='salary', height=0.5)
plt.show()
```

```python
sns.displot(data=df, x='salary', hue='training level')
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x = 'division')
plt.show()
```
