#### Apply methods

```python
# .apply method of pandas help us map and apply any custom function to every row in a Series
```

```python
df = pd.read_csv('data/tips.csv')
df.head()
```

```python
# Grab last 4 digits of the CC and store it in a row
def last_four(number):
    return str(number)[-4:]
```

```python
last_four(1199293847)
```

```python
df['cc_last_four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
# Create a new column which shows rating based on "$" symbol on how expensive restarant is

def get_expensive_rating(price):
    if price < 10:
        return "$"
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"
```

```python
get_expensive_rating(28)
```

```python
df['expensive_score'] = df['total_bill'].apply(get_expensive_rating)
df.head()
```

### Apply on multiple columns

```python
func = lambda x,y: x + y # syntax of lambda

func(1,2)
```

```python
df.head()
```

```python
# Applying custom method on multiple cols


def tip_quality(total_bill, tip):
    if tip/total_bill >= 0.25:
        return 'Generous'
    else:
        return 'Others'
```

```python
tip_quality(16.99, 1.01)
```

```python
df['Qualilty'] = df[['total_bill', 'tip']].apply(lambda df: tip_quality(df['total_bill'], df['tip']), axis = 1)
```

```python
df.head()
```

```python
# There's a way we can make it much much faster using np.vectorize with a simpler syntax

df['FasterQuality'] = np.vectorize(tip_quality)(df['total_bill'], df['tip'])
df.head()
```

```python
# Vectorize method basically makes any normal function -> numpy aware.
# So after applying vec. broadcasting facility like numpy methods is a available in normal methods as well making it much much faster
```
