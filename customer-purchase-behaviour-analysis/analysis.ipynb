```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("zeesolver/consumer-behavior-and-shopping-habits-dataset", path = "shopping_behavior_updated.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats

df = pd.read_csv(path)
df.head()
```

```python
# Task: Test whether product category preference is independent of customer age group

sns.scatterplot(x = 'Category', y = 'Age', data = df, hue='Gender')
plt.show()
```

```python
# Lets perform a Chi-2 test to confirm if this is true (because the hist scatter plot says so)
table = pd.crosstab(df['Category'], df['Age'])
chi2, p, dof, expected = stats.chi2_contingency(table)

print(p)
# As the value is little more than 0.5, we fail to reject the hypothesis and say that, it might be related to age group also
```
