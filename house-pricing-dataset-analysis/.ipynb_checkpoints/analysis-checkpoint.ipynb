```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import kagglehub

path = kagglehub.dataset_download("yasserh/housing-prices-dataset", path="Housing.csv")
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.columns[df.isna().any()].tolist() # No null columns, so data is already cleaned
```

```python
df.describe()
```

```python
# Scatter plot check if there is any relation between area and price
sns.scatterplot(x = 'area', y = 'price', data = df, hue = 'stories')
plt.show() 
# Not really becuase there are many houses whose carpet area is less but priced higher
```

```python
# Hist to observe distribution of house prices
sns.histplot(df.price, bins = 30, kde=True)
plt.show()
```

```python
# This looks like log normal destribution, To confirm lets log it out
ln_prices = np.log(df['price'])

# re-plot again
sns.histplot(ln_prices, bins = 30, kde=True)
plt.show()

# As the destr. looks more normal now, we can confirm the above data follows log-normal pattern
```

```python
from scipy import stats

skew_value = stats.skew(df['price'])
kurt_value = stats.kurtosis(df['price'])

print(f'Skew value: {skew_value}') # +ve means right skewed
print(f'Kurt value: {kurt_value}') # heavy tail
```
